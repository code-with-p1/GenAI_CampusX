```python
import warnings
warnings.filterwarnings(action='ignore')
```

```python
from langchain_google_genai import ChatGoogleGenerativeAI
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import Chroma
from langchain_core.documents import Document
```

```python
document_1 = Document(
    page_content="LangChain helps developers build LLM applications easily."
)
document_2 = Document(
    page_content="Chroma is a vector database optimized for LLM-based search."
)
document_3 = Document(
    page_content="Embeddings convert text into high-dimensional vectors."
)
document_4 = Document(
    page_content="OpenAI provides powerful embedding models."
)

documents = [document_1, document_2, document_3, document_4]
```

```python
embedding_models = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

vectorstore = Chroma.from_documents(
    documents=documents,
    embedding=embedding_models,
    collection_name="vector_store"
)
```

```python
retriever = vectorstore.as_retriever(search_type="mmr", search_kwargs={'k':2})
```

```python
query = "What is langchain ?"
results = retriever.invoke(query)
```

```python
for i, doc in enumerate(results, start=1):
    print(f"\n--- Result {i} ---")
    print(doc.page_content)
```
